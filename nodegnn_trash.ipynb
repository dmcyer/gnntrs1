```python
# =部分=6.2 神经网络结构 改造1
import torch
import torch.nn as nn
import torch.nn.functional as F
import dgl
from dgl.nn import GraphConv

debug = 0
hg_list = []


class ImprovedGCN(nn.Module):
    def __init__(self, in_feats, h_feats, num_strategies):
        super(ImprovedGCN, self).__init__()
        self.h_feats = h_feats
        
        h_feats_list = [36, 36, 36, 48, 48, 53, 57, 64, 64, 64]
        self.convs = nn.ModuleList([
            GraphConv(in_feats, h_feats_list[0]),
            GraphConv(h_feats_list[0], h_feats_list[1]),
            GraphConv(h_feats_list[1], h_feats_list[2]),
            GraphConv(h_feats_list[2], h_feats_list[3]),
            GraphConv(h_feats_list[3], h_feats_list[4]),
            GraphConv(h_feats_list[4], h_feats_list[5]),
            GraphConv(h_feats_list[5], h_feats_list[6]),
            GraphConv(h_feats_list[6], h_feats_list[7]),
            GraphConv(h_feats_list[7], h_feats_list[8]),
            GraphConv(h_feats_list[8], h_feats_list[9])
        ])
        self.fc1 = nn.Linear(h_feats, h_feats * 2)
        self.fc2 = nn.Linear(h_feats * 2, h_feats * 2)
        self.fc3 = nn.Linear(h_feats * 2, num_strategies * 2)
        # self.dropout = nn.Dropout(0.5)
        self.num_strategies = num_strategies
        self.leaky_relu = nn.LeakyReLU(0.01)

    def forward(self, g, in_feat):
        h = in_feat
        for conv in self.convs:
            h = self.leaky_relu(conv(g, h))
            
        
        # hg = h.sum(dim=0)
            
        hg = h[0]  # 只取第0号点的特征
        if debug == 1: # 调试1
            # print(hg)
            # print(hg.shape)
            hg_list.append(hg.detach().cpu().numpy())

        
        
        hg = hg.unsqueeze(0)  # 添加一个维度，变成 [1, 64]
        
        # torch.Size([1, 64])
        # torch.Size([64])
        
        # input()
        
        h = self.leaky_relu(self.fc1(hg))
        # h = self.dropout(h)
        h = self.leaky_relu(self.fc2(h))
        # h = self.dropout(h)
        out = self.fc3(h)
        
        # 分离概率和时间预测
        probs = torch.sigmoid(out[:, :self.num_strategies])
        times = F.relu(out[:, self.num_strategies:])
        
        return probs, times

# 创建模型实例
in_feats = 33  # 输入特征维度
h_feats = 64  # 增加隐藏层维度
num_strategies = 24  # 假设有24个策略

```

```python
# =部分7 是准备模型训练的环境。通过将模型、损失函数和优化器都移动到同一设备
model = ImprovedGCN(in_feats, h_feats, num_strategies)
# 重新初始化模型参数
def init_weights(m):
    if isinstance(m, nn.Linear):
        nn.init.xavier_uniform_(m.weight)
        nn.init.zeros_(m.bias)
    elif isinstance(m, GraphConv):
        nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
        if m.bias is not None:
            nn.init.zeros_(m.bias)
    # print(1) 多次初始化

model.apply(init_weights)
print("模型参数已随机初始化")

optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

criterion_prob = nn.BCELoss()  # 二元交叉熵损失用于概率
criterion_time = nn.MSELoss()  # 均方误差损失用于时间
# 检查是否有可用的GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'使用设备: {device}')


# 将损失函数移动到GPU
criterion_prob = criterion_prob.to(device)
criterion_time = criterion_time.to(device)

# 将优化器中的参数移动到GPU
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

print(f'模型、损失函数和优化器已移动到 {device}')


# 将模型移动到GPU
model = model.to(device)

```
